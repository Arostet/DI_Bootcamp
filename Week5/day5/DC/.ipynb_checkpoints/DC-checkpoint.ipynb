```python
#Number of female and male passengers
sex_counts = titanic['Sex'].value_counts()

#using these counts to plot a pie chart
labels = sex_counts.index
sizes = sex_counts.values
colors = ['lightblue', 'lightcoral']
explode = (0.1, 0)  # explode the first slice for emphasis

plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140);  
```

```python
plt.hist(titanic["Age"].dropna(), edgecolor="black")
```

```python
survived = titanic['Survived'].value_counts()
survived
```

```python
labels = survived.index
labels
number = survived.values
number
plt.bar(labels, number, color=['blue', 'green'])
plt.title('Survival Counts on the Titanic')
plt.xlabel('Survived')
plt.ylabel('Number of Passengers')
plt.xticks(labels, ['Not Survived', 'Survived'])  ;
```

```python
grouped_by_gender = titanic.groupby(['Sex', 'Survived']).size()

grouped_by_gender.plot(kind='bar')
```

```python
plt.scatter(titanic["Age"], titanic["Fare"])
```

```python
titanic['Age'].fillna(titanic['Age'].median(), inplace=True)
titanic
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
titanic['Fare'] = scaler.fit_transform(titanic[['Fare']])
titanic['Fare']
```

```python
titanic = pd.get_dummies(titanic, columns=['Sex'])
titanic
```

```python
survival_rate_gender = titanic.groupby('Sex')['Survived'].mean()
survival_rate_gender
```

```python
survival_rate_class = titanic.groupby('Pclass')['Survived'].mean()
survival_rate_class
```

```python
highest_fare_passenger = titanic['Fare'].max()
highest_fare_passenger
```
